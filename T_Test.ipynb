# T-Test

```python
import numpy as np
from scipy import stats
```

```python
sample=np.array([172,174,168,169,171,173,175,170,169,172])
sample_mean=np.mean(sample)
population_mean=170
sample_std=np.std(sample, ddof = 1)
n=len(sample)
```

```python
t_score=(sample_mean - population_mean) / (sample_std / np.sqrt(n))
t_score
```

```python
p_value = 2 * (1 - stats.t.cdf(abs(t_score),df = n-1 ))
p_value
```

```python
alpha = 0.05

if p_value < alpha :
    print("I will reject the hypothesis")
else:
    print("I will accept the hypothesis")
```
