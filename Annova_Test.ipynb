```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import f_oneway
```

```python
df=sns.load_dataset('titanic')
```

```python
df
```

```python
df = df[['age','pclass']].dropna()
```

```python
df['pclass'].unique()
```

```python
class_1 = df[df['pclass'] == 1]['age']
class_2 = df[df['pclass'] == 2]['age']
class_3 = df[df['pclass'] == 3]['age']
class_1
```

```python
f_score, p_value = f_oneway(class_1, class_2, class_3)
```

```python
print(f_score)
print(p_value)
```

```python
alpha = 0.05

if p_value < alpha :
    print("We reject the Null Hypothesis and there is significant difference")
else :
    print("We accept the Null Hypothesis")
```
