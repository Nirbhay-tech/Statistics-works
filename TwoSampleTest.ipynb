# Two Sample T-Test 

```python
import numpy as np
from scipy import stats
```

```python
groupA=[85,88,90,92,87,85,89,91,86,88]
groupB=[82,84,80,83,81,79,78,85,84,83]
```

```python
t_score , p_value = stats.ttest_ind(groupA , groupB , equal_var = False)
```

```python
t_score
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value < alpha :
    print("I will reject the Null Hypothesis")
else:
    print("I will reject the Null Hypothesis")
```
